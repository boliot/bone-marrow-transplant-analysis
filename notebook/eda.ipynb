```python
import pandas as pd
from scipy.io import arff

# 1. Charger le fichier .arff
# CORRECTIONS : Ajout du 'r' au début et de l'extension '.arff' à la fin
path = r'C:\Users\ASUS\Downloads\bone+marrow+transplant+children\bone-marrow.arff'

data, meta = arff.loadarff(path)

# 2. Convertir en DataFrame Pandas
df = pd.DataFrame(data)

# 3. NETTOYAGE : Décoder les colonnes 'bytes'
for col in df.select_dtypes([object]):
    # On vérifie si la colonne contient des bytes avant de décoder
    df[col] = df[col].apply(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x)

# 4. Afficher les premières lignes
print(df.head())
# Configurer pandas pour afficher toutes les colonnes
pd.set_option('display.max_columns', None)

# Afficher à nouveau
df.head()
pd.set_option('display.max_rows', 186)
df.head(186)
```
